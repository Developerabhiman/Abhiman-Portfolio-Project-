```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    "rating": ["G", "PG", "PG-13", "R", "G", "PG", "PG-13", "R"],
    "gross": [100, 150, 200, 250, 300, 350, 400, 450]
}
df = pd.DataFrame(data)

# Create the swarm plot
sns.swarmplot(x="rating", y="gross", data=df)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    "rating": ["G", "PG", "PG-13", "R", "G", "PG", "PG-13", "R"],
    "gross": [100, 150, 200, 250, 300, 350, 400, 450]
}
df = pd.DataFrame(data)

# Create the swarm plot
sns.swarmplot(x="rating", y="gross", data=df, size=8, hue="rating", palette="Set2")
plt.show()
```
